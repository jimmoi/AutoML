```python
import numpy as np
import pandas as pd
from sklearn.neural_network import MLPClassifier
from sklearn.model_selection import cross_val_score 
from sklearn.preprocessing import StandardScaler, OneHotEncoder
from sklearn.impute import SimpleImputer
from sklearn.pipeline import Pipeline
from sklearn.compose import ColumnTransformer

step = []

df = pd.read_csv("dataset\iris.csv")
X = df.iloc[:, :-1]
y = df.iloc[:, [-1]]

num_cols = X.select_dtypes(include=np.number).columns.tolist()
cat_cols = X.select_dtypes(exclude=np.number).columns.tolist()

num_steps = []
num_steps.append(("impute", SimpleImputer(strategy="mean")))
num_steps.append(("Scaler", StandardScaler()))
num_pipeline = Pipeline(num_steps)

cat_steps = []
cat_steps.append(("impute", SimpleImputer(strategy="most_frequent")))
cat_steps.append(("onehot", OneHotEncoder(handle_unknown="ignore")))
cat_pipeline = Pipeline(cat_steps)

preprocessor = ColumnTransformer(
        transformers=[
            ("num", num_pipeline, num_cols),
            ("cat", cat_pipeline, cat_cols)
        ])

step.append(("Preprocessor", preprocessor))
step.append(("Model", MLPClassifier()))
model = Pipeline(step)

scores = cross_val_score(model, X, y, cv=5, scoring="accuracy", n_jobs=-1)
```

```python
import numpy as np
import pandas as pd
from sklearn.neural_network import MLPRegressor
from sklearn.model_selection import cross_val_score 
from sklearn.preprocessing import MinMaxScaler, OneHotEncoder
from sklearn.impute import SimpleImputer
from sklearn.pipeline import Pipeline
from sklearn.compose import ColumnTransformer
from sklearn.datasets import fetch_california_housing

# 1. Load Dataset
data = fetch_california_housing()
X = pd.DataFrame(data.data, columns=data.feature_names)
y = data.target

# Add a fake categorical column so your cat_pipeline gets used
X['location_type'] = np.random.choice(['Urban', 'Suburban', 'Rural'], size=len(X))

# 2. Identify Columns
num_cols = X.select_dtypes(include=np.number).columns.tolist()
cat_cols = X.select_dtypes(exclude=np.number).columns.tolist()

# 3. Build Pipelines
num_pipeline = Pipeline([
    ("impute", SimpleImputer(strategy="mean")),
    ("Scaler", MinMaxScaler())
])

cat_pipeline = Pipeline([
    ("impute", SimpleImputer(strategy="most_frequent")),
    ("onehot", OneHotEncoder(handle_unknown="ignore"))
])

preprocessor = ColumnTransformer(transformers=[
    ("num", num_pipeline, num_cols),
    ("cat", cat_pipeline, cat_cols)
])

# 4. Final Model Pipeline
# Note: Added max_iter=500 because Neural Nets often need more iterations to converge
full_pipeline = Pipeline([
    ("Preprocessor", preprocessor),
    ("Model", MLPRegressor(max_iter=500, hidden_layer_sizes=(50, 50), random_state=42))
])

# 5. Cross Validation
# Using n_jobs=-1 uses all CPU cores
scores = cross_val_score(full_pipeline, X, y, cv=5, scoring="neg_mean_squared_error", n_jobs=-1)

mse_scores = -scores
print(f"Average MSE: {mse_scores.mean():.4f}")
```
